```python
for vehicle in [
    #"car",
    "truck",
    #"bus"
    #"two-wheeler"
    #"all"
]:

    if vehicle == "car":
        df = pd.read_excel(r"/Users/romain/GitHub/carculator/dev/Input data.xlsx")
        #df = pd.read_excel(r"C:\Users\sacchi_r\Documents\GitHub\carculator\dev\Input data.xlsx")

    if vehicle == "truck":
        #df = pd.read_excel(r"C:\Users\sacchi_r\Documents\GitHub\carculator_truck\dev\Input data_truck.xlsx")
        df = pd.read_excel(r"/Users/romain/GitHub/carculator_truck/dev/Input data_truck.xlsx")
        
    if vehicle == "bus":
        df = pd.read_excel(r"/Users/romain/GitHub/carculator_bus/dev/Input data_bus.xlsx")
        #df = pd.read_excel(r"C:\Users\sacchi_r\Documents\GitHub\carculator_bus\dev\Input data.xlsx")
        
    if vehicle == "two-wheeler":
        df = pd.read_excel(r"/Users/romain/GitHub/carculator_two_wheeler/dev/Input data_tw.xlsx")

    if vehicle == "all":
        df = pd.read_excel(r"/Users/romain/GitHub/commute/dev/Input data_all.xlsx")
        
    df = df.replace(np.nan, '', regex=True)
    

    d = {}
    count = 0

    d_dist = {
        'triangular':5,
        'None':1
    }


    for i, j in df.iterrows():
        if j["parameter"]!="":


            category = j["category"]

            if j["powertrain"] == "all":

                if vehicle == "car":
                    powertrain = ["PHEV-c-d", "PHEV-e", "ICEV-d", "ICEV-g", "BEV-depot",
                                  "FCEV", "HEV-d", "PHEV-d", "ICEV-p", "PHEV-c-p", "PHEV-p", "HEV-p"]

                if vehicle == "truck":
                    powertrain = ["PHEV-c-d", "PHEV-e", "ICEV-d", "ICEV-g", "BEV", "FCEV", "HEV-d", "PHEV-d"]

                if vehicle == "bus":
                    powertrain = ["ICEV-d", "ICEV-g", "HEV-d",  "BEV-opp", "BEV-motion", "BEV-depot", "FCEV"]
                    
                if vehicle == "two-wheeler":
                    powertrain = ["Human", "ICEV-p", "BEV"]
                
                

            else:
                powertrain = [x.strip() for x in j["powertrain"].split(",")]

            if j["sizes"] == "all":

                if vehicle == "car":
                    size = ["Mini", "Small", "Lower medium", "Medium", "Large", "Van", "Medium SUV", "Large SUV", "Micro"]

                if vehicle == "truck":
                    size = ["3.5t", "7.5t", "18t", "26t", "32t", "40t", "60t"]

                if vehicle == "bus":
                    size = ["9m", "13m-city", "13m-coach", "13m-city-double", "13m-coach-double", "18m"]
                    
                if vehicle == "two-wheeler":
                    size = [
                        "Kick-scooter",
                        "Bicycle <25",
                        "Bicycle cargo",
                        "Bicycle <45",
                        "Moped <4kW",
                        "Scooter <4kW",
                        "Scooter 4-11kW",
                        "Motorcycle 4-11kW",
                        "Motorcycle 11-35kW",
                        "Motorcycle >35kW",
                    ]

            else:
                size = [x.strip() for x in j["sizes"].split(",")]

            param = j["parameter"]
            unit = j["unit"]
            importance = j["importance"]
            status = j["status"]
            source = j["source"]
            comment = j["comment"]

            distr = d_dist[j["uncertainty distribution"]]

            list_years = [2000, 2010, 2020, 2030, 2040, 2050]

            for y in list_years:
                if j[y] != "":
                    name = str(count) + "-" + str(y) + "-" + param

                    if j[str(y)+".1"] !="" and j[str(y)+".1"] !="":
                        d[name]  = {
                            "amount":j[y],
                            "category":category,
                            "powertrain":powertrain,
                            "sizes":size,
                            "year":y,
                            "name":param,
                            "unit":unit,
                            "importance":importance,
                            "source":source,
                            "comment":comment,
                            "status":status,
                            "kind": "distribution",
                            "uncertainty_type":5,
                            "loc":j[y],
                            "minimum":j[str(y)+".1"],
                            "maximum":j[str(y)+".2"]
                        }
                    else:
                        d[name]  = {
                            "amount":j[y],
                            "category":category,
                            "powertrain":powertrain,
                            "sizes":size,
                            "year":y,
                            "name":param,
                            "unit":unit,
                            "importance":importance,
                            "source":source,
                            "comment":comment,
                            "status":status,
                            "kind": "distribution",
                            "uncertainty_type":1,
                            "loc":j[y]
                        }

                    count+=1

    if vehicle == "car":

        with open(r"/Users/romain/GitHub/carculator/carculator/data/default_parameters.json", 'w') as fp:
            json.dump(d, fp, indent=4)
        
        #with open(r"C:\Users\sacchi_r\Documents\GitHub\carculator\carculator\data\default_parameters.json", 'w') as fp:
        #    json.dump(d, fp, indent=4)

    if vehicle == "truck":

        #with open(r"C:\Users\sacchi_r\Documents\GitHub\carculator_truck\carculator_truck\data\default_parameters.json", 'w') as fp:
        #    json.dump(d, fp, indent=4)
            
        with open(r"/Users/romain/GitHub/carculator_truck/carculator_truck/data/default_parameters.json", 'w') as fp:
            json.dump(d, fp, indent=4)
            
        
    if vehicle == "bus":

        with open(r"/Users/romain/GitHub/carculator_bus/carculator_bus/data/default_parameters.json", 'w') as fp:
            json.dump(d, fp, indent=4)
            
        #with open(r"C:\Users\sacchi_r\Documents\GitHub\carculator_bus\carculator_bus\data\default_parameters.json", 'w') as fp:
        #    json.dump(d, fp, indent=4) 
            
    if vehicle == "two-wheeler":

        with open(r"/Users/romain/GitHub/carculator_two_wheeler/carculator_two_wheeler/data/default_parameters.json", 'w') as fp:
            json.dump(d, fp, indent=4)
            
    if vehicle == "all":
        with open(r"/Users/romain/GitHub/commute/commute/data/default_parameters.json", 'w') as fp:
            json.dump(d, fp, indent=4)
```
